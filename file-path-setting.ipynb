# 3. 여러 파일을 한번에 로드하는 방법

```python
file_csv = glob('data/store/소상*.csv')
file_csv
```

```python
glob('data/store/*_*_*.csv')
```

```python
#pd.read_csv(file_csv[0])
```

```python
file_list =[]
for file_csv_name in file_csv:
    temp = pd.read_csv(file_csv_name, low_memory=False)
    file_list.append(temp)
```

```python
#file_list에는 여러 데이터 frame이 담겨있음
len(file_list)
```

```python
#concat으로 병합
df = pd.concat(file_list)
df.shape #tuple 반환
```
